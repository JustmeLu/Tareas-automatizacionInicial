## <font color='264CC7'> Introducción </font>

Este cuaderno está diseñado practicar en el entrenamiento y evaluación de un modelo de regresión lineal utilizando datos reales. Los objetivos principales son:

- Entrenar un modelo de regresión lineal.
- Evaluar el ajuste del modelo.


Del conjunto de datos disponible [aquí](https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv), se desea predecir la variable `price` en función de las variables `bedrooms`, `bathrooms`, `sqft_living`, `sqft_lot`, `floors`.

```python
# Paquetes necesarios
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, KBinsDiscretizer, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

---
## <font color='264CC7'> Lectura de Datos </font>


Realiza la lectura de los datos y presenta las primeras filas del conjunto de datos.

```python
# Lectura de los datos desde la URL
url = "https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv"
data = pd.read_csv(url)

# Mostrar las primeras filas del conjunto de datos
print("Primeras 10 filas del dataset:")
data.head(10)
```

```python
url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv'
df = pd.read_csv(url)
# Seleccionar las columnas de interés
columns = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','price']
df = df.loc[:, columns]
df.head(10)
```

---
## <font color='264CC7'> Preprocesamiento de datos </font>

Realiza un escalado de los datos, prueba diferentes técnicas y selecciona la que mejor se ajuste a los datos.

```python
X = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']]
Y = df['price']
```

```python
# Normalización por el máximo
x_max = X.max()
X_max_normalized = X / x_max
print("Normalización por el máximo:\n")
display(X_max_normalized.head())

# Normalización Min-Max
min_max_scaler = MinMaxScaler()
X_min_max_normalized = pd.DataFrame(min_max_scaler.fit_transform(X), columns=X.columns)
print("Normalización Min-Max:\n")
display(X_min_max_normalized.head())

# Estandarización
standard_scaler = StandardScaler()
X_standardized = pd.DataFrame(standard_scaler.fit_transform(X), columns=X.columns)
print("Estandarización:\n")
display(X_standardized.head())
```

---
## <font color='264CC7'> Características y etiquetas </font>

Separa las características y las etiquetas y mostrar la dimensión de cada elemento.

```python


# Separar Características (X) y Etiqueta (Y)
features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']
X = data[features]
y = data['price']

# Mostrar dimensiones
print(f"Dimensiones de las Características (X): {X.shape}")
print(f"Dimensiones de las Etiquetas (Y): {Y.shape}")
```

---
## <font color='264CC7'> Conjunto de entrenamiento y prueba </font>

Divide los datos en dos conjuntos: uno para entrenamiento y otro para prueba y visualiza la dimensión de cada conjunto. Usar `random_state` igual al últmo número de tu cédula.

```python
# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Revisa la distribución de las etiquetas en los conjuntos de entrenamiento y prueba.

```python
# Reviso las dimensiones de los conjuntos de datos
print('Dimensiones de los conjuntos de datos:')
print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('X_test:', X_test.shape)
print('y_test:', y_test.shape)
```

---
## <font color='264CC7'> Entrenamiento </font>

Define un modelo de regresión lineal y entrenalo con los datos de entrenamiento.

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Visualizar los coeficientes
print('Ecuación de la recta:')
print(f'y = {modelo.intercept_:.2f} + {modelo.coef_[0]:.2f} * bedrooms + {modelo.coef_[1]:.2f} * bathrooms + {modelo.coef_[2]:.2f} * sqft_living + {modelo.coef_[3]:.2f} * sqft_lot + {modelo.coef_[4]:.2f} * floors')
```

---
## <font color='264CC7'> Evaluación </font>

Realiza la predicciones con el conjunto de prueba y muestra los valores en una tabla.

```python
# Predicciones del modelo
y_pred = modelo.predict(X_test)

# Comparar las predicciones con los datos de prueba
df_predicciones = pd.DataFrame({
    'Actual': y_test.values.flatten(),
    'Predicción': y_pred.flatten()
})
display(df_predicciones)
```

Calcula el error cuadrático medio.

```python
# Métricas de evaluación
mse = mean_squared_error(y_test, y_pred)

print('Error cuadrático medio:', mse)
```

```python
from sklearn.preprocessing import MaxAbsScaler, RobustScaler, Normalizer, QuantileTransformer, PowerTransformer

mse_results = {}
```

```python
max_abs_scaler = MaxAbsScaler()
X_scaled = max_abs_scaler.fit_transform(X)

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

modelo_maxabs = LinearRegression()
modelo_maxabs.fit(X_train_scaled, y_train)
y_pred_maxabs = modelo_maxabs.predict(X_test_scaled)
mse_maxabs = mean_squared_error(y_test, y_pred_maxabs)
mse_results['MaxAbsScaler'] = mse_maxabs

print(f"MSE for MaxAbsScaler: {mse_maxabs}")
```

```python
min_max_scaler = MinMaxScaler()
X_scaled_minmax = min_max_scaler.fit_transform(X)

X_train_minmax, X_test_minmax, y_train_minmax, y_test_minmax = train_test_split(X_scaled_minmax, y, test_size=0.2, random_state=0)

modelo_minmax = LinearRegression()
modelo_minmax.fit(X_train_minmax, y_train_minmax)
y_pred_minmax = modelo_minmax.predict(X_test_minmax)
mse_minmax = mean_squared_error(y_test_minmax, y_pred_minmax)
mse_results['MinMaxScaler'] = mse_minmax

print(f"MSE for MinMaxScaler: {mse_minmax}")
```

```python
robust_scaler = RobustScaler()
X_scaled_robust = robust_scaler.fit_transform(X)

X_train_robust, X_test_robust, y_train_robust, y_test_robust = train_test_split(X_scaled_robust, y, test_size=0.2, random_state=0)

modelo_robust = LinearRegression()
modelo_robust.fit(X_train_robust, y_train_robust)
y_pred_robust = modelo_robust.predict(X_test_robust)
mse_robust = mean_squared_error(y_test_robust, y_pred_robust)
mse_results['RobustScaler'] = mse_robust

print(f"MSE for RobustScaler: {mse_robust}")
```

```python
standard_scaler = StandardScaler()
X_scaled_standard = standard_scaler.fit_transform(X)

X_train_standard, X_test_standard, y_train_standard, y_test_standard = train_test_split(X_scaled_standard, y, test_size=0.2, random_state=0)

modelo_standard = LinearRegression()
modelo_standard.fit(X_train_standard, y_train_standard)
y_pred_standard = modelo_standard.predict(X_test_standard)
mse_standard = mean_squared_error(y_test_standard, y_pred_standard)
mse_results['StandardScaler'] = mse_standard

print(f"MSE for StandardScaler: {mse_standard}")
```

```python
normalizer = Normalizer()
X_scaled_normal = normalizer.fit_transform(X)

X_train_normal, X_test_normal, y_train_normal, y_test_normal = train_test_split(X_scaled_normal, y, test_size=0.2, random_state=0)

modelo_normal = LinearRegression()
modelo_normal.fit(X_train_normal, y_train_normal)
y_pred_normal = modelo_normal.predict(X_test_normal)
mse_normal = mean_squared_error(y_test_normal, y_pred_normal)
mse_results['Normalizer'] = mse_normal

print(f"MSE for Normalizer: {mse_normal}")
```

```python
quantile_transformer = QuantileTransformer(output_distribution='normal', random_state=0)
X_scaled_quantile = quantile_transformer.fit_transform(X)

X_train_quantile, X_test_quantile, y_train_quantile, y_test_quantile = train_test_split(X_scaled_quantile, y, test_size=0.2, random_state=0)

modelo_quantile = LinearRegression()
modelo_quantile.fit(X_train_quantile, y_train_quantile)
y_pred_quantile = modelo_quantile.predict(X_test_quantile)
mse_quantile = mean_squared_error(y_test_quantile, y_pred_quantile)
mse_results['QuantileTransformer'] = mse_quantile

print(f"MSE for QuantileTransformer: {mse_quantile}")
```

```python
power_transformer = PowerTransformer()
X_scaled_power = power_transformer.fit_transform(X)

X_train_power, X_test_power, y_train_power, y_test_power = train_test_split(X_scaled_power, y, test_size=0.2, random_state=0)

modelo_power = LinearRegression()
modelo_power.fit(X_train_power, y_train_power)
y_pred_power = modelo_power.predict(X_test_power)
mse_power = mean_squared_error(y_test_power, y_pred_power)
mse_results['PowerTransformer'] = mse_power

print(f"MSE for PowerTransformer: {mse_power}")
```

```python
print("Summary of Mean Squared Error for different scalers:")
for scaler_name, mse_value in mse_results.items():
    print(f"- {scaler_name}: {mse_value}")
```




- **Escalado RobustScaler:** Este escalador utiliza el rango intercuartílico (IQR) y la mediana para escalar los datos, lo que lo hace robusto frente a los valores atípicos (outliers). Es útil cuando el conjunto de datos contiene muchos outliers que podrían afectar negativamente a otros métodos de escalado basados en la media y la desviación estándar.

- **Escalado Normalizer:** A diferencia de otros escaladores que operan sobre las características (columnas), `Normalizer` escala cada muestra (fila) individualmente para que tenga una norma unitaria (longitud 1). Es especialmente útil cuando se utilizan métricas basadas en la distancia angular o la similitud de coseno, ya que preserva la dirección de los vectores de las muestras.

- **Escalado QuantileTransformer:** Este transformador mapea las características a una distribución uniforme o normal utilizando la función cuantil. Esto significa que los valores atípicos se distribuyen de manera más uniforme o se acercan a una distribución gaussiana, lo que puede ser beneficioso para modelos que asumen una distribución normal o son sensibles a outliers. Puede ser útil para transformar datos con distribuciones muy sesgadas.

- **Escalado PowerTransformer:** Este transformador aplica una transformación de potencia a cada característica para hacer que su distribución sea más gaussiana (similar a una distribución normal). Incluye las transformaciones Box-Cox y Yeo-Johnson. Hacer que los datos se ajusten más a una distribución normal puede mejorar el rendimiento de muchos modelos lineales y basados en distancia.
